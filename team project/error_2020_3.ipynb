```python
import pandas as pd
import matplotlib.pyplot as plt
import platform
import koreanize_matplotlib
```

```python
file = '../team_data/경영애로사항_2020.xlsx'
data = pd.read_excel(file, index_col = [0,1], header = [0], skiprows=1)
data
```

```python
d = data.xs('권역별', level = 0)
d 
```

```python
d = d.T
d = d.drop(['모집단 크기 (개)'], axis = 0)
d = d.head(20)
d = d.replace('-',0)
```

```python
d
```

```python
xlist = list(d.columns)
ylist = list(d.index)
```

```python
def graph(city, c):
    data = d[city]
    x = d.index
    y = d.values
    plt.figure(figsize = (20,10))
    data.plot(kind = 'bar', grid = True, rot = 30, color = c)
    plt.ylim(0, 60)
    plt.title(city, fontsize = 'xx-large')
    plt.show()
   
def graph2(city, c):
    data = d[city]
    x = data.index
    y = data.values
    y = sorted(y, reverse = True)
    
    plt.figure(figsize = (20,10))
    plt.bar(x, y, color=c)
    plt.xticks(rotation=300)
    plt.grid(axis = 'y')
    #data.plot(kind = 'bar', grid = True, rot = 30, color = c)
    plt.ylim(0, 60)
    plt.title(city, fontsize = 'xx-large')
    plt.show()
    
    
def graph3(city, c):
    data = d[city]
    data = data.sort_values(ascending=False)
    x = data.index
    y = data.values
    
    plt.figure(figsize = (20,10))
    plt.bar(x, y, color=c)
    plt.xticks(rotation=300)
    plt.grid(axis = 'y')
    #data.plot(kind = 'bar', grid = True, rot = 30, color = c)
    plt.ylim(0, 60)
    plt.title(city, fontsize = 'xx-large')
    plt.show()
```

```python
graph3('서울', 'orange')
```
