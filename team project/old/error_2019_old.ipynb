- 2019 분석

```python
file = '../team_data/경영애로사항_2019.xlsx'
data = pd.read_excel(file, index_col = [0,1], header = [0], skiprows=1)
data
```

```python
d = data.xs('권역별', level = 0)
d
```

```python
d.shape
```

```python
for i in range(d.shape[0]):
    for j in range(d.shape[1]):
        d.iloc[i,j].replace('-', 0)
d
```

```python
ylist = list(d.columns)
ylist
```

```python
len(ylist)
```

```python
xlist = list(d.index)
xlist
```

```python
value = d[ylist[1]]
value
```

```python
plt.figure(figsize = (10, 5))
plt.bar(xlist, d[ylist[1]])
plt.title('자금확보')
```

```python
value = d[ylist[2]]
plt.figure(figsize = (10, 5))
plt.bar(xlist, d[ylist[2]])
plt.title(ylist[2])
plt.show
```

```python
plt.figure(figsize = (20, 10))

for num in range(1, 6):
    plt.subplot(1,6,num)
    value = d[ylist[num]]
    plt.bar(xlist, d[ylist[num]])
    plt.title(ylist[num])
plt.show()
```

## 여기서부터 시각화

```python
def graph (start, end, c):
    plt.figure(figsize = (20,10))
    
    for num in range(start, end+1):
        num1 = num - start + 1
        plt.subplot(1, 5, num1)
        plt.bar(xlist, d[ylist[num]], color = c)
        plt.ylim(0, 50)
        plt.title(ylist[num])
        #xticks를 rotation하고 싶은데 그건 어떻게 하는 거지?
        
plt.show()
```

```python
graph(1, 5, 'blue')
```

```python
graph(6,10,'blue')
```

```python
graph(11,15,'blue')
```

```python
graph(16,20,'blue')
```

```python
graph(21,22,'blue')
```
