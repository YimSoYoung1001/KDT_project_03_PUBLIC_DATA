```python
def graph(city, c):
    data = d[city]
    x = d.index
    y = d.values
    plt.figure(figsize = (20,10))
    data.plot(kind = 'bar', grid = True, rot = 30, color = c)
    plt.ylim(0, 60)
    plt.title(city, fontsize = 'xx-large')
    plt.show()
    
    
def graph2(city, c):
    data = d[city]
    x = data.index
    y = data.values
    y = sorted(y, reverse = True)
    
    plt.figure(figsize = (20,10))
    plt.bar(x, y, color=c)
    plt.xticks(rotation=300)
    plt.grid(axis = 'y')
    #data.plot(kind = 'bar', grid = True, rot = 30, color = c)
    plt.ylim(0, 60)
    plt.title(city, fontsize = 'xx-large')
    plt.show()
```

```python
graph2('서울', 'blue')
```
