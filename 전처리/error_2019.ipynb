- 2019 분석

```python
file = '../team_data/경영애로사항_2019.xlsx'
data = pd.read_excel(file, index_col = [0,1], header = [0], skiprows=1)
data
```

```python
d = data.xs('권역별', level = 0)
d
```

```python
d.shape
```

```python
d = d.T
d
```

```python
d = d.drop(['모집단 크기 (개)'], axis = 0)
```

```python
d.shape
```

```python
d = d.head(20)
d
```

```python
d = d.replace('-',0)
```

```python
xlist = list(d.columns)
```

```python
ylist = list(d.index)
```

```python
def graph(city, c):
    data = d[city]
    x = d.index
    y = d.values
    plt.figure(figsize = (20,10))
    data.plot(kind = 'bar', grid = True, rot = 30, color = c)
    plt.ylim(0, 60)
    plt.title(city, fontsize = 'xx-large')
    plt.show()
```

```python
def graph2(city):
    data = d[city]
    x = d.index
    y = d.values
    plt.figure(figsize = (20,10))
    data.plot(kind = 'pie')
    plt.title(city, fontsize = 'xx-large')
    plt.show()
```

```python
graph('서울', 'green')
```

```python
graph2('서울')
```

```python
def graph (start, end, c):
    plt.figure(figsize = (20,10))
    
    for num in range(start, end+1):
        num1 = num - start + 1
        plt.subplot(1, 5, num1)
        plt.bar(xlist, d[ylist[num]], color = c)
        plt.ylim(0, 50)
        plt.title(ylist[num])
        #xticks를 rotation하고 싶은데 그건 어떻게 하는 거지?
        
plt.show()
```

```python
graph(1, 5, 'blue')
```

```python
graph(6,10,'blue')
```

```python
graph(11,15,'blue')
```
