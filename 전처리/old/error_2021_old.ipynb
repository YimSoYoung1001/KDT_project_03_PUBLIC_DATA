```python
import pandas as pd
import matplotlib.pyplot as plt
import platform
import koreanize_matplotlib
```

```python
file = '../team_data/경영애로사항_2021.xlsx'
data = pd.read_excel(file, index_col = [0,1], header = [0], skiprows=1)
data
```

```python
d = data.xs('권역별', level = 0)
d
```

```python
ylist = list(d.columns)
xlist = list(d.index)
```

```python
len(ylist)
```

```python
def graph (start, end, c):
    plt.figure(figsize = (20,10))
    
    for num in range(start, end+1):
        num1 = num - start + 1
        plt.subplot(1, 5, num1)
        plt.bar(xlist, d[ylist[num]], color = c)
        plt.ylim(0, 50)
        plt.title(ylist[num])
        #xticks를 rotation하고 싶은데 그건 어떻게 하는 거지?
        
plt.show()
```

```python
graph(1,5,'green')
```

```python
graph(6,10,'green')
```

```python
graph(11,15,'green')
```

```python
graph(16,18,'green')
```
