```python
import pandas as pd
import matplotlib.pyplot as plt
import platform
import koreanize_matplotlib
```

```python
file = '../team_data/경영애로사항_2020.xlsx'
data = pd.read_excel(file, index_col = [0,1], header = [0], skiprows=1)
data
```

```python
d = data.xs('권역별', level = 0)
d 
```

```python
d = d.T
d = d.drop(['모집단 크기 (개)'], axis = 0)
d = d.head(20)
d = d.replace('-',0)
```

```python
d
```

```python
xlist = list(d.columns)
ylist = list(d.index)
```

```python
def graph(city, c):
    data = d[city]
    x = d.index
    y = d.values
    plt.figure(figsize = (20,10))
    data.plot(kind = 'bar', grid = True, rot = 30, color = c)
    plt.ylim(0, 60)
    plt.title(city, fontsize = 'xx-large')
    plt.show()
```

```python
graph('서울', 'orange')
```

```python
graph('인천/경기', 'orange')
```

```python
graph('대전/세종/충청/강원', 'orange')
```

```python
graph('부산/울산/경남', 'orange')
```

```python
graph('대구/경북', 'orange')
```

```python
graph('광주/전라/제주', 'orange')
```

```python
def graph (start, end, c):
    plt.figure(figsize = (20,10))
    
    for num in range(start, end+1):
        num1 = num - start + 1
        plt.subplot(1, 5, num1)
        plt.bar(xlist, d[ylist[num]], color = c)
        plt.ylim(0, 50)
        plt.title(ylist[num])
        #xticks를 rotation하고 싶은데 그건 어떻게 하는 거지?
        
plt.show()
```

```python
graph(1,5,'orange')
```
